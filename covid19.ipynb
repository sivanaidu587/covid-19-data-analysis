```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

# Sample dataset
np.random.seed(42)
data = pd.DataFrame({
    "x": np.arange(1, 101),
    "y": np.random.randint(10, 100, 100),
    "category": np.random.choice(["A","B","C","D"], 100),
    "value": np.random.randn(100)
})

# 1. Line Chart
plt.figure()
plt.plot(data["x"], data["y"])
plt.title("Line Chart")
plt.show()

# 2. Bar Charts (8 examples)
for i in range(8):
    plt.figure()
    plt.bar(data["category"].value_counts().index,
            data["category"].value_counts().values)
    plt.title(f"Bar Chart {i+1}")
    plt.show()

# 3. Horizontal Bar Charts (4)
for i in range(4):
    plt.figure()
    plt.barh(data["category"].value_counts().index,
             data["category"].value_counts().values)
    plt.title(f"Horizontal Bar Chart {i+1}")
    plt.show()

# 4. Pie Charts (4)
for i in range(4):
    plt.figure()
    plt.pie(data["category"].value_counts().values,
            labels=data["category"].value_counts().index,
            autopct='%1.1f%%')
    plt.title(f"Pie Chart {i+1}")
    plt.show()

# 5. Scatter Plots (5)
for i in range(5):
    plt.figure()
    plt.scatter(data["x"], data["y"], c=data["value"], cmap="viridis")
    plt.title(f"Scatter Plot {i+1}")
    plt.show()

# 6. Histograms (2)
for i in range(2):
    plt.figure()
    plt.hist(data["value"], bins=20, color="skyblue", edgecolor="black")
    plt.title(f"Histogram {i+1}")
    plt.show()

# 7. Box Plots (2)
for i in range(2):
    plt.figure()
    sns.boxplot(x="category", y="value", data=data)
    plt.title(f"Box Plot {i+1}")
    plt.show()

# 8. Area Charts (3)
for i in range(3):
    plt.figure()
    plt.fill_between(data["x"], data["y"], alpha=0.5)
    plt.title(f"Area Chart {i+1}")
    plt.show()


# 9. Heatmaps (3)
for i in range(3):
    plt.figure()
    # Only select numeric columns
    corr = data.select_dtypes(include=[np.number]).corr()
    sns.heatmap(corr, annot=True, cmap="coolwarm")
    plt.title(f"Heatmap {i+1}")
    plt.show()


# 10. Bubble Charts (2)
for i in range(2):
    plt.figure()
    sizes = np.clip((data["value"]+2)*50, 10, 500)  # safe sizes
    plt.scatter(data["x"], data["y"], s=sizes, alpha=0.5, c=data["value"], cmap="coolwarm")
    plt.title(f"Bubble Chart {i+1}")
    plt.show()


# 11. Violin Plots (2)
for i in range(2):
    plt.figure()
    sns.violinplot(x="category", y="value", data=data)
    plt.title(f"Violin Plot {i+1}")
    plt.show()

# 12. KDE Plots (2)
for i in range(2):
    plt.figure()
    sns.kdeplot(data["value"], fill=True)
    plt.title(f"KDE Plot {i+1}")
    plt.show()

# 13. Radar Charts (2)
labels = ["Metric1","Metric2","Metric3","Metric4","Metric5"]
stats = np.random.randint(10,100,5)
angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
stats = np.concatenate((stats,[stats[0]]))
angles += angles[:1]
for i in range(2):
    fig = plt.figure()
    ax = fig.add_subplot(111, polar=True)
    ax.plot(angles, stats, 'o-', linewidth=2)
    ax.fill(angles, stats, alpha=0.25)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)
    plt.title(f"Radar Chart {i+1}")
    plt.show()

# 14. Treemaps (2) - using plotly
for i in range(2):
    fig = px.treemap(data, path=["category"], values="y")
    fig.show()

# 15. Sunburst Charts (2)
for i in range(2):
    fig = px.sunburst(data, path=["category"], values="y")
    fig.show()

# 16. Choropleth Maps (2) - sample with country codes
world_data = pd.DataFrame({
    "country":["USA","IND","BRA","FRA","DEU"],
    "cases":[100,80,60,40,30]
})
for i in range(2):
    fig = px.choropleth(world_data, locations="country",
                        color="cases", locationmode="ISO-3")
    fig.show()

# 17. Animated Time-Series (2)
for i in range(2):
    fig = px.scatter(data, x="x", y="y", animation_frame="x",
                     size="y", color="category")
    fig.show()

# 18. Sankey Diagrams (2)
for i in range(2):
    fig = go.Figure(go.Sankey(
        node=dict(label=["Start","A","B","End"]),
        link=dict(source=[0,0,1,2], target=[1,2,3,3], value=[8,4,2,8])
    ))
    fig.show()

# 19. Parallel Coordinates (2)
for i in range(2):
    fig = px.parallel_coordinates(data, dimensions=["x","y","value"],
                                  color="y")
    fig.show()

# 20. Interactive Dashboards (5) - using plotly
for i in range(5):
    fig = px.scatter(data, x="x", y="y", color="category",
                     facet_col="category")
    fig.show()
```
